Exploratory Data Analysis

```python
# Importing Libraries
import ast
import pandas as pd
import matplotlib.pyplot as plt  

# Loading Data
df = pd.read_csv(r"C:\Users\SAMA\Downloads\data_jobs.csv")

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA_US=df[(df['job_title_short']=='Data Analyst')&(df['job_country']=='United States')]
```

```python
df_plot=df_DA_US['job_location'].value_counts().sort_values(ascending=False).head(10).to_frame()
```

```python
import seaborn as sns
```

```python
sns.barplot(df_plot,x='count',y=df_plot.index,hue='count',palette='crest',legend=False)
sns.set_theme(style='ticks')
sns.despine()
plt.title('Count of Job Location for Data Analyst in the US')
plt.ylabel('')
plt.xlabel('Number of Jobs')
plt.show()
```

```python
my_dict={'job_work_from_home':'Work From Home'
        ,'job_no_degree_mention': 'Job Degree Req'
        ,'job_health_insurance': 'Health Insurance Offered'}
```

```python
fig,ax=plt.subplots(1,3)
fig.set_size_inches((10,5))

for i,(column,title) in enumerate(my_dict.items()):
    
    df_plot=df_DA_US[column].value_counts()
    # df_plot.plot(kind='pie',startangle=90,ax=ax[i],autopct='%1.1f%%')
    df_plot.plot(
    kind='pie',
    startangle=90,
    ax=ax[i],
    autopct='%1.1f%%',
    textprops={'fontsize':12}
)

    ax[i].set_title(title, fontsize=15)

   
    
fig.suptitle('')    
fig.tight_layout()    
```

```python
top_companies=df_DA_US['company_name'].value_counts().sort_values(ascending=False).head(10).to_frame()
```

```python
sns.set_theme(style='ticks')
sns.barplot(top_companies,x='count',y='company_name',hue='count',palette='crest')
sns.despine()
plt.title('Count of Companies for Data Analyst in the US',fontsize=17)
plt.ylabel('')
plt.xlabel('Number of Jobs')
plt.legend().set_visible(False)



plt.show()
```
